```python
import numpy as np
import matplotlib.pyplot as plt
```

### Just Graphs

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
plt.plot(x,y, label='The line', linewidth = 2, marker = '.', markersize = 10, markerfacecolor = 'red', markeredgecolor = 'black', color = "green")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
plt.plot(x,y, label='The line', linewidth = 2, linestyle = '--', marker = '.', markersize = 20, markerfacecolor = 'red', markeredgecolor = 'black', color = "green")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
plt.plot(x,y, 'b^--')
plt.show()
```

#### How to make an equation and draw it's Graph

```python
x = np.linspace(0,100,10000)
y = x**2 + x*4 + 9
plt.figure(dpi=500,figsize=(10,6))
plt.plot(x,y, linewidth = 2)
plt.savefig('Graph.svg', dpi = 3000)
plt.show()
```

### Bars

```python
x = ['A', 'B', 'C', 'D', 'E', 'H']
y = [4,8,3,7,10, 8]
bar = plt.bar(x,y)
bar[0].set_hatch('/')
bar[1].set_hatch('o')
bar[2].set_hatch('-')
bar[3].set_hatch('O')
bar[4].set_hatch('*')
bar[5].set_hatch('+')
plt.figure(figsize=(20,15), dpi= 1000)
plt.show()
```
