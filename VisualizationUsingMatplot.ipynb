```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3], dtype='float64')
y = np.array([2,4,6], dtype='float64')
z = np.array([2,3,5], dtype='float64')
plt.title("On the graph", fontdict={'fontsize':20, 'color':'blue'})
plt.xlabel('V1', color= 'red')
plt.ylabel('V2', color= 'red')
plt.plot(x,y, label= 'Yousif', color='red', linewidth=2, marker='^',linestyle='--', markersize = 10, markeredgecolor= 'green')
plt.plot(y,z, label= 'nour', color='blue', linewidth=2.5)
plt.xticks(np.arange(1,7, step= 0.5))
plt.yticks(np.arange(1,7, step=1))
plt.legend()
plt.show()
```
