##### Building a function that convert the two lists to dictionary
* Note: The two lists numst have the same size

```python
def list_to_dict(list1, list2):
  if (len(list1) != len(list2)):
    print ("The two lists are not equal in size")
    return -1
  else:
    a = dict.fromkeys(list1)
    for b, c in zip(list1, list2):
      a[b] = c
    return a

#best in performance
def list_to_dict2(list1, list2):
  if (len(list1) != len(list2)):
    print ("The two lists are not equal in size")
    return -1
  else:
    return {a: b for a, b in zip(list1, list2)}
```

```python
print (list_to_dict([n for n in range(31)], [n**2 for n in range(31)]))
print (list_to_dict2([n for n in range(31)], [n**2 for n in range(31)]))
print (list_to_dict2([n for n in range(31)], [n**2 for n in range(30)]))
```

```python
# making a function that returns a random decimal number in a range
def rand_decimal_nums(a,b):
  fraction = rn.random()
  if (fraction == 0):
    return rn.randint(a,b)
  else:
    num = rn.randint(a,b)
    if (num == b):
      num -= 1
    return float(num) + rn.random()
```
